```python
import pandas as pd
import glob
import os

# Read and concatenate all CSV files in the directory
df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "*.csv"))))

# Check if 'Unnamed: 0' exists and drop it if present
if 'Unnamed: 0' in df.columns:
    df = df.drop(['Unnamed: 0'], axis=1)

# Reset index after concatenation
all_df = df.reset_index(drop=True)

# Now 'all_df' contains the cleaned DataFrame
```
